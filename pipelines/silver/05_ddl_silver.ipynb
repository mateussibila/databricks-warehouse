```sql
-- ===========================================================================
-- Silver Layer - Create Tables
-- ===========================================================================
-- This script creates all Silver tables for the Medallion Architecture
-- Existing tables will be dropped if they exist
-- ===========================================================================

-- -------------------------------
-- CRM Customers
-- -------------------------------
-- Drop tables if they exist
DROP TABLE IF EXISTS silver.crm_cust_info;
CREATE TABLE silver.crm_cust_info (
    cst_id              INT,
    cst_key             STRING,
    cst_firstname       STRING,
    cst_lastname        STRING,
    cst_marital_status  STRING,
    cst_gndr            STRING,
    cst_create_date     DATE,
    dwh_create_date     TIMESTAMP
);

-- -------------------------------
-- CRM Products
-- -------------------------------
-- Drop tables if they exist
DROP TABLE IF EXISTS silver.crm_prd_info;
CREATE TABLE silver.crm_prd_info (
    prd_id              INT,
    cat_id              STRING, -- new column created
    prd_key             STRING,
    prd_nm              STRING,
    prd_cost            INT,
    prd_line            STRING,
    prd_start_dt        DATE, -- was originaly TIMESTAMP but data doesn't contain hours
    prd_end_dt          DATE, -- was originaly TIMESTAMP but data doesn't contain hours
    dwh_create_date     TIMESTAMP
);

-- -------------------------------
-- CRM Sales Details
-- -------------------------------
-- Drop tables if they exist
DROP TABLE IF EXISTS silver.crm_sales_details;
CREATE TABLE silver.crm_sales_details (
    sls_ord_num         STRING,
    sls_prd_key         STRING,
    sls_cust_id         INT,
    sls_order_dt        DATE, -- previously INT
    sls_ship_dt         DATE, -- previously INT
    sls_due_dt          DATE, -- previously INT
    sls_sales           INT,
    sls_quantity        INT,
    sls_price           INT,
    dwh_create_date     TIMESTAMP
);

-- -------------------------------
-- ERP Locations
-- -------------------------------
-- Drop tables if they exist
DROP TABLE IF EXISTS silver.erp_loc_a101;
CREATE TABLE silver.erp_loc_a101 (
    cid                 STRING,
    cntry               STRING,
    dwh_create_date     TIMESTAMP
);

-- -------------------------------
-- ERP Customers
-- -------------------------------
-- Drop tables if they exist
DROP TABLE IF EXISTS silver.erp_cust_az12;
CREATE TABLE silver.erp_cust_az12 (
    cid                 STRING,
    bdate               DATE,
    gen                 STRING,
    dwh_create_date     TIMESTAMP
);

-- -------------------------------
-- ERP Product Categories
-- -------------------------------
-- Drop tables if they exist
DROP TABLE IF EXISTS silver.erp_px_cat_g1v2;
CREATE TABLE silver.erp_px_cat_g1v2 (
    cat                 STRING,
    subcat              STRING,
    maintenance         STRING,
    dwh_create_date     TIMESTAMP
);
```
